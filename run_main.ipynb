```python
# A notebook that can be used to load and prep aCurve data from in .csv format. 
# The prep_csv function removes every third column (empty) and creates a pandas dataframe
# with time stamps as indices and variable names as in row 2 in the .csv file. 

# The data is then split into training and testing data using the split function. Data is
# scaled (normalized) using the function scale. 

# Several machine learning models are trained and evaluated based on MAE, MSE, RMSE and R2. 
# The predictions are plotted in two ways. 
```

```python
# --- IMPORTING OWN FUNCTIONS ---
from prep_csv import prep_csv
from pre_process import split, scale
from evaluation import evaluate_models, plot_predictions
```

```python
# --- RUN MAIN FUNCTION ---
file_path = "test_data.csv"

# LOAD AND PREP the data
data_red, time_stamp = prep_csv(file_path)

# SPLIT the data
X_train, X_test, y_train, y_test = split(data_red,0.2)

# PRE-PROCESS the data
X_train, X_test = scale(X_train,X_test)

# TRAIN models
results, predictions = evaluate_models(X_train, X_test, y_train, y_test)

# EVALUATE models
for model_name, metrics in results.items():
    print(f"\n{model_name} Results:")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")

# PLOT predictions vs actual data
plot_predictions(y_test, predictions)
```
